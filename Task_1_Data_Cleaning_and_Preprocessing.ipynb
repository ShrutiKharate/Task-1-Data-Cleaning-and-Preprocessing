```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/food_coded.csv')
```

```python
print("Initial dataset shape:", df.shape)
df.head(10)
```

```python
df.info()
```

```python
# Check missing values
missing = df.isnull().sum()
print("\nMissing values per column:\n", missing)
```

```python
# Fill remaining nulls (for numeric and object types separately)
df = df.fillna({
    col: df[col].median() if df[col].dtype != 'O' else df[col].mode()[0]
    for col in df.columns
})
```

```python
# Remove Duplicate Rows
df = df.drop_duplicates()
print("\nShape after removing duplicates:", df.shape)
```

```python
# Convert Gender values to 'Female' and 'Male'
gender_map = {1: 'Female', 2: 'Male'}
df['Gender'] = df['Gender'].map(gender_map)
```

```python
# Fix Column Names (lowercase, no spaces)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# GPA and weight should be float or int
df['gpa'] = pd.to_numeric(df['gpa'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')

# Re-check types
df.dtypes
```

```python
# Print summary
print("\n--- Data Cleaning Summary ---")
print(f"Final shape: {df.shape}")
print(f"Remaining missing values:\n{df.isnull().sum().sum()} total")
print(f"Column names: {list(df.columns)}")
```

```python
df.to_csv('/content/food_coded_cleaned.csv', index=False)
print("Cleaned dataset saved as 'food_coded_cleaned.csv'")
```

```python
# Summary of Changes:

# Dropped columns with >50% missing values.
# Filled missing values using median/mode based on type.
# Removed duplicate rows.
# Standardized Gender column values using a mapping.
# Cleaned column headers to lowercase with underscores.
# Converted GPA and weight to numeric types.
```
